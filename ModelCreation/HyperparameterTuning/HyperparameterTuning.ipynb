```python
import json
import matplotlib.pyplot as plt
import numpy as np

learning_rate = 9e-5
epochs = 400
batch_sizes = [4, 8, 16, 32]
weight_decays = [0.01, 0.001, 0.0001]
training_steps = [10000, 20000, 50000]

# initialize empty lists to accumulate training loss values for each configuration
all_train_loss_values = []
all_train_batch_size_values = []

for weight_decay in weight_decays:
    for batch_size in batch_sizes:
        
        train_loss_values = []
        train_batch_size_values = []

        for training_step in training_steps:

            # load metrics from file
            metrics_file = f"train_logs_lr_{learning_rate}_epochs_{epochs}_batch_{batch_size}_decay_{weight_decay}_steps_{training_step}.json"
            with open(metrics_file, "r") as f:
                metrics = json.load(f)

            # extract training loss and batch size
            train_loss = [log["train_loss"][0] for log in metrics]
            train_batch_size = [log["train_samples"] for log in metrics]
    
            # append to arrays
            train_loss_values.append(train_loss)
            train_batch_size_values.extend(train_batch_size)

        # append training loss values to the cumulative lists
        all_train_loss_values.append(train_loss_values)
        all_train_batch_size_values.append(train_batch_size_values)

        # print final training loss value for this configuration
        print(f"Training Loss for Batch Size {batch_size}, Weight Decay {weight_decay}, Training Steps {training_step}: {train_loss[-1]}")

# plot all training loss values in a single plot
plt.figure()
for i, weight_decay in enumerate(weight_decays):
    for j, batch_size in enumerate(batch_sizes):
        # extract the accumulated training loss values for this configuration
        train_loss_values = all_train_loss_values[i*len(batch_sizes)+j]
        train_batch_size_values = all_train_batch_size_values[i*len(batch_sizes)+j]
        # plot the training loss values
        plt.plot(train_batch_size_values, train_loss_values, label=f"Batch Size {batch_size}, Weight Decay {weight_decay}")
plt.title("Training Loss for all Configurations")
plt.xlabel("Training Samples")
plt.ylabel("Training Loss")
plt.legend()
plt.show()
```

```python
import json
import matplotlib.pyplot as plt
import numpy as np

learning_rate = 9e-5
epochs = 400
batch_sizes = [4, 8, 16, 32]
weight_decays = [0.01, 0.001, 0.0001]
training_steps = [10000, 20000, 50000]

# initialize empty lists to accumulate training loss values for each configuration
all_train_loss_values = []
all_train_batch_size_values = []
all_train_step_values = []

for weight_decay in weight_decays:
    for batch_size in batch_sizes:
        
        train_loss_values = []
        train_batch_size_values = []
        train_step_values = []

        for training_step in training_steps:

            # load metrics from file
            metrics_file = f"train_logs_lr_{learning_rate}_epochs_{epochs}_batch_{batch_size}_decay_{weight_decay}_steps_{training_step}.json"
            with open(metrics_file, "r") as f:
                metrics = json.load(f)

            # extract training loss, batch size, and training step
            train_loss = [log["train_loss"][0] for log in metrics]
            train_batch_size = [log["train_samples"] for log in metrics]
            train_steps = [log["epoch"]*len(train_loss)/epochs*training_step for log in metrics]
    
            # append to arrays
            train_loss_values.append(train_loss)
            train_batch_size_values.extend(train_batch_size)
            train_step_values.extend(train_steps)

        # append training loss and step values to the cumulative lists
        all_train_loss_values.append(train_loss_values)
        all_train_batch_size_values.append(train_batch_size_values)
        all_train_step_values.append(train_step_values)

        # print final training loss value for this configuration
        print(f"Training Loss for Batch Size {batch_size}, Weight Decay {weight_decay}, Training Steps {training_step}: {train_loss[-1]}")

# plot all training loss values in a single plot
plt.figure()
for i, weight_decay in enumerate(weight_decays):
    for j, batch_size in enumerate(batch_sizes):
        # extract the accumulated training loss and step values for this configuration
        train_loss_values = all_train_loss_values[i*len(batch_sizes)+j]
        train_batch_size_values = all_train_batch_size_values[i*len(batch_sizes)+j]
        train_step_values = all_train_step_values[i*len(batch_sizes)+j]
        # plot the training loss values
        plt.plot(train_step_values, train_loss_values, label=f"Batch Size {batch_size}, Weight Decay {weight_decay}")
plt.title("Training Loss for all Configurations")
plt.xlabel("Training Steps")
plt.ylabel("Training Loss")
plt.legend()
plt.show()
```
